```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
# --- 1. Load and Prepare the Data ---

# Set a nice style for our plots
sns.set_theme(style="whitegrid")

file_name = 'satellite_data.json'
print(f"Loading data from '{file_name}'...")

try:
    # Load the JSON file into a pandas DataFrame
    df = pd.read_json(file_name)
    print(f"Successfully loaded {len(df)} satellite records.")
except FileNotFoundError:
    print(f"Error: The file '{file_name}' was not found in this directory.")
    print("Please make sure you have uploaded 'satellite_data.json' to Colab's session storage.")
    exit()
```

```python
# --- 2. Data Cleaning and Preparation ---

# The LAUNCH_DATE is currently a string. We need to convert it to a
# proper date format so we can extract the year.
df['LAUNCH_DATE'] = pd.to_datetime(df['LAUNCH_DATE'], errors='coerce')

df['LAUNCH_YEAR'] = df['LAUNCH_DATE'].dt.year

df.dropna(subset=['LAUNCH_YEAR'], inplace=True)

df['LAUNCH_YEAR'] = df['LAUNCH_YEAR'].astype(int)

print("Data preparation complete.")
```

```python
# --- 3. Create Visualizations ---

# --- Plot 1: Number of Satellites per Country ---
plt.figure(figsize=(10, 6))

country_plot = sns.countplot(x='COUNTRY', data=df, order=df['COUNTRY'].value_counts().index, palette='viridis')

country_plot.set_title('Total Number of Satellites by Country', fontsize=16)
country_plot.set_xlabel('Country Code', fontsize=12)
country_plot.set_ylabel('Number of Satellites', fontsize=12)

for p in country_plot.patches:
    country_plot.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center', xytext=(0, 9), textcoords='offset points')

plt.show()
```
